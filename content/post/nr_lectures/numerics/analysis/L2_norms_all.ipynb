```julia
using PyPlot
using LaTeXStrings
using DelimitedFiles
pygui(false)
```

```julia
# give the directory where the data from all the runs are saved
your_dir = "/home/thanasis/repos/my_chaos/presentations/nr_lectures_nottingham_group/lec4/numerics/examples/runs_center/"
```

```julia
##LOAD SMOOTH DATA SH##

# DATA DIR; change to the directory where you have the norms of the SH_smooth_B1.jl model
smooth_data_dir_SH = your_dir*"SH_smooth/norms_self_2"

#N=0 & N=1 & N=2
L2_0_1_c0     = readdlm(smooth_data_dir_SH*"/L2_0_1_c0.dat", comments=true)[:,2]
L2_1_2_c0     = readdlm(smooth_data_dir_SH*"/L2_1_2_c0.dat", comments=true)[:,2]
smooth_p_0_SH = log2.(L2_0_1_c0./L2_1_2_c0) 
smooth_t_0_SH = readdlm(smooth_data_dir_SH*"/L2_0_1_c0.dat", comments=true)[:,1]

#N=1 & N=2 & N=3
L2_1_2_c1     = readdlm(smooth_data_dir_SH*"/L2_1_2_c1.dat", comments=true)[:,2]
L2_2_3_c1     = readdlm(smooth_data_dir_SH*"/L2_2_3_c1.dat", comments=true)[:,2]
smooth_p_1_SH = log2.(L2_1_2_c1./L2_2_3_c1) 
smooth_t_1_SH = readdlm(smooth_data_dir_SH*"/L2_1_2_c1.dat", comments=true)[:,1]

#N=2 & N=3 & N=4
L2_2_3_c2     = readdlm(smooth_data_dir_SH*"/L2_2_3_c2.dat", comments=true)[:,2]
L2_3_4_c2     = readdlm(smooth_data_dir_SH*"/L2_3_4_c2.dat", comments=true)[:,2]
smooth_p_2_SH = log2.(L2_2_3_c2./L2_3_4_c2) 
smooth_t_2_SH = readdlm(smooth_data_dir_SH*"/L2_2_3_c2.dat", comments=true)[:,1]

#N=3 & N=4 & N=5
L2_3_4_c3     = readdlm(smooth_data_dir_SH*"/L2_3_4_c3.dat", comments=true)[:,2]
L2_4_5_c3     = readdlm(smooth_data_dir_SH*"/L2_4_5_c3.dat", comments=true)[:,2]
smooth_p_3_SH = log2.(L2_3_4_c3./L2_4_5_c3) 
smooth_t_3_SH = readdlm(smooth_data_dir_SH*"/L2_3_4_c3.dat", comments=true)[:,1];
```

```julia
## PLOT ##

fig = figure("pyplot_subplot_touching",figsize=(17,5.5))
subplots_adjust(hspace=0.08, wspace=0.04, top = 0.99, bottom = 0.05, right=0.98, left=0.07)# Set the vertical spacing between axes

# Strongly hyperbolic smooth data
subplot(111) # Create the 1st fig in 2 rows and 2 columns
grid("on")
#x-axis
#xlim(0.0,1.0) # Set the x-limits
#xticks(0.0:0.2:1.0,fontsize="0") # Set the x-tick range and step size
xlabel("time",fontsize="0")
#y-axis
ylabel(L"\mathcal{C}",fontsize="20")
#yticks(0.0:1.0:3.0,fontsize="15")
#ylim(0.0,3.0)
#plot
ti = 1
tf = 257
plot(smooth_t_0_SH[ti:tf], smooth_p_0_SH[ti:tf], linestyle=":" , linewidth=3,  
label=L"\log_2{\left(||\mathbf{u}_{h_0}-\mathbf{u}_{h_1}||\,/\,||\mathbf{u}_{h_1}-\mathbf{u}_{h_2}||\right)}")
plot(smooth_t_1_SH[ti:tf], smooth_p_1_SH[ti:tf], linestyle="-.", linewidth=2.5, 
label=L"\log_2{\left(||\mathbf{u}_{h_1}-\mathbf{u}_{h_2}||\,/\,||\mathbf{u}_{h_2}-\mathbf{u}_{h_3}||\right)}")
plot(smooth_t_2_SH, smooth_p_2_SH, linestyle="--", linewidth=2.5,
label=L"\log_2{\left(||\mathbf{u}_{h_2}-\mathbf{u}_{h_3}||\,/ \,||\mathbf{u}_{h_3}-\mathbf{u}_{h_4}||\right)}")
plot(smooth_t_3_SH, smooth_p_3_SH, linestyle="-" , linewidth=2.2,
label=L"\log_2{\left(||\mathbf{u}_{h_3}-\mathbf{u}_{h_4}||\,/ \,||\mathbf{u}_{h_4}-\mathbf{u}_{h_5}||\right)}")
# LEGEND
legend(fontsize="15", loc=1)
################################

fig.canvas.draw() # Update the figure
```
